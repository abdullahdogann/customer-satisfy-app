```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_data.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.rename(columns={'product_photos_qty':'photos_count'}, inplace= True)
df.rename(columns= {'different_between_approved_purchase' : 'time_different_hour'}, inplace = True)
df.rename(columns = {'delivery' : 'delivered_on_time'}, inplace = True)
df
```

```python
order_status = {
    0 : 'delivered',
    1 : 'canceled'
}
```

```python
df['order_status'] = df.order_status.map(order_status)
```

```python
df.payment_type.value_counts()
```

```python
payment_type = {
    1 : 'credit_cart',
    2 : 'coupon',
    3 : 'voucher',
    4 : 'debit cart'
}
```

```python
df['payment_type'] = df.payment_type.map(payment_type)
```

```python
review_score = {
    0 : 'satisfied',
    1 : 'not satisfied'
}
```

```python
df['review_score'] = df.review_score.map(review_score)
```

```python
price = {
    1 : '<50',
    2 : '50 - 100',
    3 : '100 - 250',
    4 : '250 - 500',
    5 : '500 - 1000',
    6 : '1000 - 2000',
    7 : '2000 - 3000',
    8 : '3000 - 4000',
    9 : '4000 - 5000',
    10 : '>5000'
}
```

```python
df['price'] = df.price.map(price)
```

```python
def photos_count(x):
    if x <= 5:
        return 1
    elif x > 5 and x <= 10:
        return 2
    elif x > 10 and x <= 15:
        return 3
    else:
        return 4
```

```python
df['photos_count'] = df.photos_count.apply(photos_count)
```

```python
photos_count = {
    1 : '<5',
    2 : '5 - 10',
    3 : '10 -15',
    4 : '>15'
}
```

```python
df['photos_count'] = df.photos_count.map(photos_count)
```

```python
time_different_hour = {
    1 : '<1',
    2 : '1 - 5',
    3 : '5 - 10',
    4 : '10 - 20',
    5 : '20 - 30',
    6 : '30 - 40',
    7 : '40 - 50',
    8 : '>50'
}
```

```python
df['time_different_hour'] = df.time_different_hour.map(time_different_hour)
```

```python
delivery_on_time = {
    0 : 'No',
    1 : 'Yes'
}
```

```python
df['delivered_on_time'] = df.delivered_on_time.map(delivery_on_time)
```

```python
delivered_hour = {
    1 : 'Morning',
    2 : 'Midday',
    3 : 'Evening',
    4 : 'Night'
}
```

```python
df['delivered_hour'] = df.delivered_hour.map(delivered_hour)
```

```python
df['delivered_season'] = df['delivered_season'].apply(lambda x: 'Winter' if x == 1 else 'Summer')
```

```python
seller_count = {
    1 : '<10',
    2 : '10 - 50',
    3 : '50 - 250',
    4 : '250 - 500',
    5 : '500 - 1000',
    6 : '1000 - 2000',
    7 : '2000 - 3000',
    8 : '3000 - 4000',
    9 : '4000 - 5000',
    10 : '>5000'
}
```

```python
df['seller_count'] = df.seller_count.map(seller_count)
```
